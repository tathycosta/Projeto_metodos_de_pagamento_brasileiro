Importando as bibliotecas

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Carregando o Dataset

```python
brazilian_payment_methods=pd.read_csv('https://docs.google.com/spreadsheets/d/e/2PACX-1vTaTchZzSNPqCie19TykRfZgqC1ie0zu73H9j75NVpxa6gRsmE9KMMW6fLUgk5x18sEPDO01Bz0xv_P/pub?gid=1553293454&single=true&output=csv')
```

```python
display(brazilian_payment_methods.head(5))
```

Troca dos Nomes das Colunas para Português

```python
brazilian_payment_methods=brazilian_payment_methods.rename(columns={
    'YearMonth':'Mês e Ano',
    'quantityPix':'Número de transações Pix',
    'valuePix':'Valor total das transações Pix',
    'quantityTED':'Número de transações TED',
    'valueTED':'Valor total das transações TED',
    'quantityTEC':'Número de transações TEC',
    'valueTEC':'Valor total das transações TEC',
    'quantityBankCheck':' Número de transações de cheque',
    'valueBankCheck':'Valor total das transações de cheque',
    'quantityBrazilianBoletoPayment':'Número de transações de boleto',
    'valueBrazilianBoletoPayment':'Valor total das transações de boleto',
    'quantityDOC':'Número de transações DOC',
    'valueDOC':'Valor total das transações DOC'
    })

display(brazilian_payment_methods.head(5))
```

Seleção de Colunas

```python
brazilian_payment_methods_mes_ano_numero_valor=brazilian_payment_methods[['Mês e Ano','Número de transações Pix','Valor total das transações Pix']]
display(brazilian_payment_methods_mes_ano_numero_valor.head(10))
```

Filtragem de Dados

```python
valor_total_pix=brazilian_payment_methods[brazilian_payment_methods['Valor total das transações Pix']>1000000]
display(valor_total_pix)
```

Criação de Nova Coluna

```python
brazilian_payment_methods['Valor médio por transação Pix']=(
brazilian_payment_methods['Valor total das transações Pix'] /
brazilian_payment_methods['Número de transações Pix']
)
display(brazilian_payment_methods[['Mês e Ano','Número de transações Pix','Valor total das transações Pix','Valor médio por transação Pix']].head(5))
```

Agrupamento de Dados (Groupby)

```python
brazilian_payment_methods['Mês e Ano'] = pd.to_datetime(brazilian_payment_methods['Mês e Ano'], format="%Y%m")
brazilian_payment_methods['Ano'] = brazilian_payment_methods['Mês e Ano'].dt.year
brazilian_payment_methods['Mês'] = brazilian_payment_methods['Mês e Ano'].dt.month
soma_total_dos_valores_pix = brazilian_payment_methods.groupby('Ano')['Valor total das transações Pix'].sum().reset_index()
soma_total_dos_valores_pix.sort_values(by='Valor total das transações Pix', ascending=False)
```

```python
brazilian_payment_methods['Mês']
```

Ordenação de Dados

```python
transacoes_ordenadas = brazilian_payment_methods.sort_values(by='Valor total das transações Pix', ascending=False)
top_10_transacoes = transacoes_ordenadas.head(10)
display(top_10_transacoes)
```

Contagem de Dados

```python
soma_mensal = brazilian_payment_methods.groupby('Mês e Ano')['Valor total das transações Pix'].sum()
meses_com_mais_de_1M = (soma_mensal > 1000000)
contagem_meses = meses_com_mais_de_1M.value_counts().reset_index()
print(f"\nNúmero total de meses com mais de 1.000.000 de transações PIX: {contagem_meses['count'][1]}")
```

Média e Mediana

```python
media_transacoes_TED = np.mean(brazilian_payment_methods['Valor total das transações TED'])
print('Média de Valor total de transações TED:', media_transacoes_TED)
mediana_transacoes_TED = np.median(brazilian_payment_methods['Valor total das transações TED'])
print('Mediana de Valor total de transações TED:',mediana_transacoes_TED)
```

Agrupamento e agregação

```python
agrupamento_boleto_ano = brazilian_payment_methods.groupby('Ano')['Valor total das transações de boleto'].mean().reset_index()
agrupamento_boleto_ano.sort_values(by='Valor total das transações de boleto', ascending=False)
```

Visualização de dados

```python
plt.plot(brazilian_payment_methods['Mês e Ano'], brazilian_payment_methods['Valor total das transações Pix'], linestyle = '-')
plt.title('Evolução do valor das transações Pix por ano')
plt.xlabel('Mês e Ano')
plt.ylabel('Valor total das transações Pix')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Substituição de valores

```python
brazilian_payment_methods['Número de transações DOC'] = brazilian_payment_methods['Número de transações DOC'].replace(0.0, np.nan)
display(brazilian_payment_methods['Número de transações DOC'].head())
```

Correlação entre Métodos de pagamento

```python
brazilian_payment_methods_correlacao = brazilian_payment_methods
brazilian_payment_methods_correlacao = brazilian_payment_methods_correlacao.set_index('Mês e Ano')
correlacao = brazilian_payment_methods_correlacao['Valor total das transações Pix'].corr(brazilian_payment_methods_correlacao['Valor total das transações TED'])
print(f'Como os valores de correlação foram {correlacao} proximos de +1 significa que quando os valores PIX aumentam os de TED também tedem a aumentar')
```

Comparação de Crescimento entre PIX e TED

```python
brazilian_payment_methods['crescimento_pix'] = brazilian_payment_methods['Valor total das transações Pix'].pct_change() * 100
brazilian_payment_methods['crescimento_ted'] = brazilian_payment_methods['Valor total das transações TED'].pct_change() * 100

plt.plot(brazilian_payment_methods['Mês e Ano'], brazilian_payment_methods['crescimento_pix'], linestyle = '-', label= 'Crescimento % Pix', color = 'blue')
plt.plot(brazilian_payment_methods['Mês e Ano'], brazilian_payment_methods['crescimento_ted'], linestyle = '-', label= 'Crescimento % TED', color = 'orange')

plt.title('Crescimento Percentual das Transações Pix e TED')
plt.xlabel('Ano')
plt.ylabel('Crescimento Percentual (%)')

plt.grid(True)

plt.legend()

plt.xticks(rotation=45)
plt.show()
```

Participação de Cada Método de Pagamento

```python
ultimo_mes = brazilian_payment_methods['Mês e Ano'].max()
ultimo_mes_data = brazilian_payment_methods[brazilian_payment_methods['Mês e Ano'] == ultimo_mes].iloc[0]
valor_total = ultimo_mes_data[['Valor total das transações Pix', 'Valor total das transações TED', 'Valor total das transações TEC','Valor total das transações de boleto', 'Valor total das transações DOC']].sum()
participacao_percentual = (ultimo_mes_data[['Valor total das transações Pix', 'Valor total das transações TED', 'Valor total das transações TEC','Valor total das transações de boleto', 'Valor total das transações DOC']] / valor_total) * 100
print(f"Participacao percentual no ultimo mês {ultimo_mes.strftime('%Y-%m')}:")
print(participacao_percentual)
```

Valor Médio por Transação

```python
brazilian_payment_methods['Mês e Ano'] = pd.to_datetime(brazilian_payment_methods['Mês e Ano'])
ultimo_ano = brazilian_payment_methods['Mês e Ano'].dt.year.max()
ultimo_mes = brazilian_payment_methods[brazilian_payment_methods['Mês e Ano'].dt.year == ultimo_ano]['Mês e Ano'].dt.month.max()
dados_ultimo_mes = brazilian_payment_methods[(brazilian_payment_methods['Mês e Ano'].dt.year == ultimo_ano) & (brazilian_payment_methods['Mês e Ano'].dt.month == ultimo_mes)]

dados_ultimo_mes['Valor médio por transação Pix'] = (dados_ultimo_mes['Valor total das transações Pix'] / dados_ultimo_mes['Número de transações Pix'])
dados_ultimo_mes['Valor médio por transação TED'] = (dados_ultimo_mes['Valor total das transações TED'] / dados_ultimo_mes['Número de transações TED'])
dados_ultimo_mes['Valor médio por transação TEC'] = (dados_ultimo_mes['Valor total das transações TEC'] / dados_ultimo_mes['Número de transações TEC'])
dados_ultimo_mes['Valor médio por transação boleto'] = (dados_ultimo_mes['Valor total das transações de boleto'] / dados_ultimo_mes['Número de transações de boleto'])
dados_ultimo_mes['Valor médio por transação DOC'] = (dados_ultimo_mes['Valor total das transações DOC'] / dados_ultimo_mes['Número de transações DOC'])

display(dados_ultimo_mes[[
    'Mês e Ano',
    'Valor médio por transação Pix',
    'Valor médio por transação TED',
    'Valor médio por transação TEC',
    'Valor médio por transação boleto',
    'Valor médio por transação DOC'
]])
```

Valor Total por Método de Pagamento

```python
brazilian_payment_methods['Mês e Ano'] = pd.to_datetime(brazilian_payment_methods['Mês e Ano'])
ultimo_ano = brazilian_payment_methods['Mês e Ano'].dt.year.max()
ultimo_mes = brazilian_payment_methods[brazilian_payment_methods['Mês e Ano'].dt.year == ultimo_ano]['Mês e Ano'].dt.month.max()

dados_ultimo_mes = brazilian_payment_methods[
    (brazilian_payment_methods['Mês e Ano'].dt.year == ultimo_ano) &
    (brazilian_payment_methods['Mês e Ano'].dt.month == ultimo_mes)
]

valores_totais = {
    'PIX': dados_ultimo_mes['Valor total das transações Pix'].sum(),
    'TED': dados_ultimo_mes['Valor total das transações TED'].sum(),
    'TEC': dados_ultimo_mes['Valor total das transações TEC'].sum(),
    'Boleto': dados_ultimo_mes['Valor total das transações de boleto'].sum(),
    'DOC': dados_ultimo_mes['Valor total das transações DOC'].sum()
}

metodos = list(valores_totais.keys())
valores = list(valores_totais.values())

plt.figure(figsize=(10, 6))
plt.bar(metodos, valores, color=['blue'])
plt.title(f'Valor Total das Transações por Método de Pagamento ({ultimo_mes}/{ultimo_ano})')
plt.xlabel('Método de Pagamento')
plt.ylabel('Valor Total das Transações')
plt.show()
```
